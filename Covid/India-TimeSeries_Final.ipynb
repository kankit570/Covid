```python
import requests
import json
import pandas as pd
import plotly.graph_objects as go
```

```python
url = 'https://covidstat.info/graphql/'

query = """query {
  country(name: "India") {
    historical {
      date
      cases
      deaths
      recovered
      todayCases
      todayRecovered
      todayDeaths
    }
  }
}"""

r = requests.post(url, json={'query': query})
```

```python
json_data = json.loads(r.text)
```

```python
df_data = json_data['data']['country']['historical']
df1 = pd.DataFrame(df_data)
```

```python
df1['active']=df1['cases']-df1['deaths']-df1['recovered']
df1['recoveryrate']=df1['recovered']/df1['cases']
```

```python
fig = go.Figure()


fig.add_trace(go.Scatter(x=df1['date'], y=df1['cases'],
                    mode='lines',
                    name='TimeSeries Confirmed Cases'))
fig.add_trace(go.Scatter(x=df1['date'], y=df1['deaths'],
                    mode='lines',
                    name='TimeSeries Deceased Cases'))
fig.add_trace(go.Scatter(x=df1['date'], y=df1['recovered'],
                    mode='lines',
                    name='TimeSeries Recovered Cases'))

fig.add_trace(go.Scatter(x=df1['date'], y=df1['active'],
                    mode='lines',
                    name='TimeSeries Active Cases'))
fig.update_layout(title='Covid-19 Cases',
                   xaxis_title='Date',
                   yaxis_title='No Of Cases')
fig.show()
```

```python
fig1 = go.Figure()


fig1.add_trace(go.Scatter(x=df1['date'], y=df1['todayCases'],
                    mode='lines',
                    name='TimeSeries Daily Confirmed Cases'))
fig1.add_trace(go.Scatter(x=df1['date'], y=df1['todayDeaths'],
                    mode='lines',
                    name='TimeSeries Daily Deceased Cases'))
fig1.add_trace(go.Scatter(x=df1['date'], y=df1['todayRecovered'],
                    mode='lines',
                    name='TimeSeries Daily Recovered Cases'))

fig1.update_layout(title='Covid-19 Daily Cases',
                   xaxis_title='No Of Cases',
                   yaxis_title='Date')
fig1.show()
```
